## **Importar librerias**

```
# Import necessary libraries
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

## **Buscar la proteína objetivo**

La búsqueda de la proteína objetivo implica localizar y seleccionar la proteína específica a la que se dirige un compuesto o una actividad biológica. En el contexto del descubrimiento de medicamentos, identificar la proteína objetivo es fundamental para comprender la interacción entre un compuesto y la función biológica que se está investigando.

Este paso es crucial para el diseño y la evaluación de posibles fármacos, ya que la interacción entre un compuesto y la proteína objetivo puede afectar directamente el resultado terapéutico. En el contexto de la base de datos ChEMBL, se pueden realizar búsquedas para identificar proteínas específicas y su relación con compuestos o actividades biológicas registradas.

### **Busqueda para aromatasa**

```
# Target search for coronavirus
target = new_client.target
target_query = target.search('aromatase')
targets = pd.DataFrame.from_dict(target_query)
targets
```

### **Seleccionar y recuperar datos de bioactividad (primer entrada en este caso)**

```
selected_target = targets.target_chembl_id[0]
selected_target
```

Aquí, vamos a recuperar solo los datos de bioactividad para "CHEMBL3927" que están reportados como valores de IC$_{50}$ en unidades de nM (nanomolares).

```
activity = new_client.activity
res = activity.filter(target_chembl_id=selected_target).filter(standard_type="IC50")
```

```
df = pd.DataFrame.from_dict(res)
```

```
df.head(4)
```

```
df.standard_type.unique()
```

Finalmente guardamos los resultados en un archivo csv llamado: **aromatase_bioactivity_data.csv**.

```
df.to_csv('aromatase_bioactivity_data.csv', index=False)
```

## **Manejo de datos faltantes**
Si algún compuesto tiene un valor faltante en la columna de **standard_value**, entonces elimínelo.

```
df2 = df[df.standard_value.notna()]
df2
```

## **Preprocesamiento de los datos de bioactividad**

### **Etiquetado de compuestos como activos, inactivos o intermedios**
Los datos de bioactividad están en la unidad de IC50. Se considerarán **activos** los compuestos que tengan valores menores a 1000 nM, mientras que aquellos con valores mayores a 10,000 nM serán considerados **inactivos**. En cuanto a los valores que estén entre 1,000 y 10,000 nM, se denominarán **intermedios**.

```
bioactivity_class = []
for i in df2.standard_value:
  if float(i) >= 10000:
    bioactivity_class.append("inactive")
  elif float(i) <= 1000:
    bioactivity_class.append("active")
  else:
    bioactivity_class.append("intermediate")
```

### **Iteramos *molecule_chembl_id* dentro de una lista**

```
mol_cid = []
for i in df2.molecule_chembl_id:
  mol_cid.append(i)
```

### **Iteramos *canonical_smiles* dentro de una lista**

```
canonical_smiles = []
for i in df2.canonical_smiles:
  canonical_smiles.append(i)
```

### **Iteramos *standard_value* dentro de una lista**

```
standard_value = []
for i in df2.standard_value:
  standard_value.append(i)
```

### **Combinacion de las 4 listas anteriores en un DataFrame**

```
data_tuples = list(zip(mol_cid, canonical_smiles, bioactivity_class, standard_value))
df3 = pd.DataFrame( data_tuples,  columns=['molecule_chembl_id', 'canonical_smiles', 'bioactivity_class', 'standard_value'])
```
